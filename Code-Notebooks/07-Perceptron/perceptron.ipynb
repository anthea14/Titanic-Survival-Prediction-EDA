Importing necessary Libraries

```python
import numpy as np
import pandas as pd
```

Activation Function

```python
def step_function(weighted_sum):
    if weighted_sum >=0:
        return 1
    else:
        return 0
```

Perceptron Prediction Function

```python
def perceptron(experience, skills, weights, bias):
    weighted_sum = (experience * weights[0]) + (skills * weights[1]) + bias
    return step_function(weighted_sum)
```

Training Function

```python
def train_perceptron(train, outputs, weights, bias, num_samples, learning_rate, epochs):
    weights = weights.copy()
    bias = bias
    for epoch in range(epochs):
        total_error = 0
        for i in range(num_samples):
            experience = train[i][0]
            skills = train[i][1]

            prediction = perceptron(experience, skills, weights, bias)
            error = outputs[i] - prediction
            total_error += error ** 2
            
            weights[0] += learning_rate * error * experience
            weights[1] += learning_rate * error * skills
            bias += learning_rate * error
        
        if total_error == 0:
            print(f"Training complete after {epoch} epochs.")
            break
    return weights, bias
```

Sample data

```python
train = [[2,2],[1,1],[3,1],[1,5],[2,4],[1,2]]
outputs =[1,0,1,1,1,0]
```

Initialize Parameters

```python
initial_weights = np.array([0.0,0.0])
initial_bias = 0.0
learning_rate = 0.1
epochs = 100
```

Training the Model

```python
trained_weights, trained_bias =train_perceptron(train, outputs, initial_weights, initial_bias, len(train), learning_rate, epochs)
```

```python
print(f"final weights: Experience = {trained_weights[0]}, Skills = {trained_weights[1]}, Bias = {trained_bias}")
```

Testing the Model

```python
print("\nTesting the trained perceptron:")
for i in range(len(train)):
    exp, skill = train[i]
    prediction = perceptron(exp, skill, trained_weights, trained_bias)
    print(f"Input: {train[i]}, Expected: {outputs[i]}, Predicted: {prediction}")
```

Prediction

```python
print(f"Input: [1, 10], Expected: {perceptron(1, 10, trained_weights, trained_bias)}")
```
