```python
import numpy as np
```

Function for Computing Slope & Intercept

```python
def LinearRegression(x, y, size):
    x_mean = np.mean(x)
    y_mean = np.mean(y)
    num = 0
    den =0
    for i in range(size):
        num += (x[i] - x_mean) * (y[i] - y_mean)
        den += (x[i] - x_mean) **2
    
    m = num/den #slope
    b = y_mean - m * x_mean #intercept
    return m, b
```

Sample Training Data

```python
x = np.array([1400, 1600, 1700, 1875, 1100])
y = np.array([245, 312, 279, 308, 199])
size = len(x)
```

Calling the Function

```python
m, b = LinearRegression(x, y, size)
```

Linear Regression Line

```python
print(f" Linear Regression Model: y = {m:.2f}x + {b:.2f}")
```

Predicting Price for new House data

```python
new_x = 1500
predicted_y = m * new_x + b
print(f" Predicted price for house size {new_x} sq ft: ${predicted_y:.2f}K")
```

Allowing User to provide the Input

```python
user_x = float(input(" Enter house size in sq ft: "))
user_predicted_y = m * user_x + b
print(f" Predicted price for house size {user_x} sq ft: ${user_predicted_y:.2f}K")
```
